# Filter TOI Data

```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
raw_df = pd.read_csv('../data/raw/TOI (TESS)/tess_object_of_interest.csv', sep=',')
raw_df.head(3)
```

```python
useful_columns = [
    'tfopwg_disp',     
    'pl_orbper',       
    'pl_trandurh',     
    'pl_trandep',      
    'pl_rade',         
    'pl_insol',        
    'pl_eqt',          
    'st_teff',         
    'st_logg',         
    'st_rad'           
]

df = raw_df[useful_columns].copy()

df.head()
```

```python
print('Size: ', df.shape, '\nColumns:', df.columns)
```

```python
new_columns_name = {
    'tfopwg_disp': 'STATUS',
    'pl_orbper': 'ORBITAL_PERIOD_DAYS',
    'pl_trandurh': 'TRANSIT_DURATION_HOURS',
    'pl_trandep': 'TRANSIT_DEPTH_PPM',
    'pl_rade': 'PLANET_RADIUS_REARTH',
    'pl_insol': 'PLANET_INSOLATION_EFLUX',
    'pl_eqt': 'PLANET_EQ_TEMP_K',
    'st_teff': 'STELLAR_TEFF_K',
    'st_logg': 'STELLAR_LOGG_CMS2',
    'st_rad': 'STELLAR_RADIUS_RSUN',
}

df = df.rename(columns=new_columns_name)
df
```
