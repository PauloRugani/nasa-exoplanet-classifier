# Data visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../data/processed/3_data_cleaning.csv', sep=';')
df.head(3)
```

```python
for i in df.columns[1:]:
    sns.violinplot(data=df, x=df[i], hue=df['STATUS'])
    plt.show()
```

```python
sns.pairplot(df, hue='STATUS', corner=True)
plt.show()
```

```python
sns.clustermap(df.corr(), cmap='coolwarm', center=0, figsize=(12, 12))
plt.show()
```
