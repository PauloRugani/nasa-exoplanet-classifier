# Merge KOI and TOI datasets

```python
import pandas as pd
```

```python
koi_df = pd.read_csv('../data/processed/1_filtered_koi_data.csv', sep=';')
koi_df.head(3)
```

```python
toi_df = pd.read_csv('../data/processed/1_filtered_toi_data.csv', sep=';')
toi_df.head(3)
```

```python
df = pd.concat([koi_df, toi_df], ignore_index=True)
df.head(3)
```

```python
df.to_csv('../data/processed/1.1_merged_data.csv', sep=';', index=False)
```
